## **Exercise 1: Detective work**

````python
import ROOT
%jsroot on 

c1 = ROOT.TCanvas("c1","c1",800,600)

f = ROOT.TF2(
    "f", #function name?
    "sin(x)*sin(y)/x*y", #Equation
    -10, 10,   # x range
    0, 10     # y range
)

f.Draw("Surf1")
c1.Draw()
````

## **Walkthrough: Working with Histograms**

````python
import ROOT

ROOT.TH1.AddDirectory(False)

c2 = ROOT.TCanvas("c2","c2",800,600)
h1 = ROOT.TH1D("hist1", "Histogram from a Gaussian", 100, -3, 3) ##100 bins from 03 to 3

##Data points need to be filled in :)
h1.FillRandom("gaus",10000)

print(h1.GetBinWidth(0)) #even w/o points bin is 0.6

#h1.Draw("e") Finds the error... 
h1.Draw()
c2.Draw()
````

## Creating our Own function :D

````python
%jsroot on
import ROOT
ROOT.TH1.AddDirectory(False)

c3 = ROOT.TCanvas("c3","c3",800,600)

##Minor function difference
myfunc = ROOT.TF1("myfunc", "gaus", -3, 3) #Equals sign different from tutorial... it's the C++ way of doing stuff :D

myfunc.SetParameters(10., 1.0, 0.5)

h2 = ROOT.TH1D("hist2", "Histogram from my function", 100, -3, 3)
h2.FillRandom("myfunc", 10000)

h2.Draw()
myfunc.Draw("same")

c3.Draw()
````

Note from tutorial
```Note that we could also set the functions parameters individually:

[] myfunc.SetParameter(1,-1.0)
[] h2.FillRandom("myfunc",10000)
```

## **Walkthrough: Fitting to a Gaussian distribution**
> I will be using my own graph
> Also Note to me!: You can use the little camera on the side to activate the stat-box

````python
import ROOT

ROOT.TH1.AddDirectory(False)

c4 = ROOT.TCanvas("c4","c4",800,600)
h3 = ROOT.TH1D("hist1", "Histogram from a Gaussian", 100, -3, 3)

h3.FillRandom("gaus", 10000)

#Fit function made :D
f = ROOT.TF1("f", "gaus", -5, 5)
# Fit histogram
h3.Fit(f, "R")

# Draw histogram with error bars
h3.Draw("E")

# Draw the fit function on top
f.SetLineColor(ROOT.kRed)
f.SetLineWidth(2)
f.Draw("same")

c4.Draw()
````
